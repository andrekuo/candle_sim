```python
##### PyAEDT Connection
from pyaedt import Hfss

version='2021.1'
proj = hfss.pinfo.project_name
designname = hfss.get_active_design_name()
solutiontype = 'DrivenModal'
setupname = em1.setup.name

hfss_pyaedt = Hfss(specified_version=version, projectname=proj, designname=designname, solution_type=solutiontype,
                setup_name=setupname, new_desktop_session=False, close_on_exit=True)

##### Update to fix silicon parameters to ultra cold temp
materials = hfss_pyaedt.materials
silicon = materials.checkifmaterialexists('silicon')
silicon.permittivity = 11.45

##### Create Setup 
# Note: you need to have an design w/ ports ot use MultiFreq & Interpolating
'Setup starts here.'
setupname = em1.setup.name
setup = hfss_pyaedt.setups[0]
setup.props['SolveType'] = 'MultiFrequency'
setup.props['MultipleAdaptiveFreqsSetup'] = OrderedDict([('6.3GHz', [0.01]), # Resonator
                                                             ('4GHz', [0.01]),   # Qubit
                                                             ('3GHz', [0.01])])  # Qubit 
setup.props['MaximumPasses'] = 2
setup.props['MinimumPasses'] = 1
setup.props['MinimumConvergedPasses'] = 1
setup.props['PercentRefinement'] = 30
setup.props['BasisOrder'] = "1"

##### Create Sweep (within setup)
sweepname = "Sweep"
def create_draft_sweep(sweepname):
    hfss_pyaedt.create_linear_count_sweep(
        setupname=setupname,
        unit="GHz",
        freqstart=5.6, #Ghz
        freqstop=7.1, #Ghz
        sweepname=sweepname,
        sweep_type="Interpolating",
        num_of_freq_points=10000,
        interpolation_max_solutions=5000,
        interpolation_tol=0.01,
        save_fields=False
    )

##### Remove previous design, used to get MultiFreq & Interpolating 
hfss.clean_active_design()
```

## Per Cavity + Qubit Setup

```python
import pandas as pd
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime
from qiskit_metal.analyses.em.transmission_fitting import fit_transmission

now = datetime.now().strftime("%Y%m%d %H%M%S")

##### Run ANSYS simulation
def run_sim(i):
    """
    Args:
        i (int): name integer of qubit
    """ 
    hfss.clean_active_design()
    
    ### Render design
    if i == 4 or i == 5 or i == 6: # Finger coupler
        selection = [f'Q{i}', f'cpw{i}', f'ncap{i}']
        open_pins = [(f'ncap{i}', 'prime_start'), (f'ncap{i}', 'prime_end')]
        port_list = [(f'ncap{i}', 'prime_start', 50), (f'ncap{i}', 'prime_end', 50)]
    else: # Hanging resonator coupler
        selection = [f'Q{i}', f'cpw{i}', f'clt{i}']
        open_pins = [(f'clt{i}', 'prime_start'), (f'clt{i}', 'prime_end')]
        port_list = [(f'clt{i}', 'prime_start', 50), (f'clt{i}', 'prime_end', 50)]
        
    hfss.clean_active_design()

    hfss.render_design(selection=selection,
                       open_pins=open_pins,
                       port_list=port_list,
                       box_plus_buffer=True)
    
    ### Preseed Mesh
    hfss.modeler.mesh_length(
                'qubit',
                [f'cross_Q{i}'],
                MaxLength='0.01mm')
    if i == 4 or i == 5 or i == 6: # Finger coupler
        hfss.modeler.mesh_length(
                    'cpw_mesh',
                    [f'trace_cpw{i}', f'c_connector_arm_Q{i}', f'prime_cpw_ncap{i}', f'cap_body_0_ncap{i}', f'cap_body_1_ncap{i}', f'second_cpw_top_ncap{i}', f'second_cpw_bottom_ncap{i}'],
                    MaxLength='3um')
    else: # Hanging resonator coupler
        hfss.modeler.mesh_length(
                    'cpw_mesh',
                    [f'trace_cpw{i}', f'c_connector_arm_Q{i}', f'prime_cpw_clt{i}', f'second_cpw_clt{i}'],
                    MaxLength='3um')
        
    ### Run draft simulation
    setup.update()
    draft_sweepname = sweepname + f'_draft_{i}'
    create_draft_sweep(sweepname=draft_sweepname)

    hfss.analyze_sweep(draft_sweepname, setupname)
        
    ### Parse data and save result
    index = "S21"
    
    freqs, Pcurves, Pparams = hfss.get_params([index])
    
    try: # If you can fit, then let's resimulate at a closer value.
        _, _ = fit_transmission(freqs,Pparams[index].values)
        
        ### Run final simulation
        # Sweep around Resonant Frequency (f_res)
        f_res = freqs[np.argmin(np.abs(Pparams[index].values))]
        freqstart = np.round(f_res/1e9,3)-0.01
        freqstop = np.round(f_res/1e9,3)+0.01
        
        final_sweepname = sweepname + f"_closer_{i}"
        
        hfss_pyaedt.create_linear_count_sweep(
            setupname=setupname,
            unit="GHz",
            freqstart=freqstart, #Ghz
            freqstop=freqstop, #Ghz
            sweepname=final_sweepname,
            sweep_type="Interpolating",
            num_of_freq_points=10000,
            interpolation_max_solutions=5000,
            interpolation_tol=0.01,
            save_fields=False
        )
        hfss.analyze_sweep(final_sweepname, setupname)
        
        
        # [THIS IS SAVED] Grab & Analyze Data
        freqs, Pcurves, Pparams = hfss.get_params([index])
            
        Pparams.to_csv(f'{now}.csv')

        plt.plot(freqs, Pparams[index].values.real, label=f'Re({index})')
        plt.plot(freqs, Pparams[index].values.imag, label=f'Im({index})')
        plt.xlabel('Freq (GHz)')
        plt.ylabel(index)
        plt.title(index)
        plt.legend()
        plt.savefig(f'{now}.png')

        try:
            fit_paras, plots = fit_transmission(freqs,Pparams[index].values) # This fits the transmission, plots the fit over the raw data, and returns only the parameters best fit parameters and plots
            scattering_package = {'data':Pparams, 'fit_paras':fit_paras, 'fit_plots':plots}
            
            return scattering_package
        except:
            print(f"WARNING: Qubit #{i} failed to fit in the FINAL round. Skipped this qubit.")
        
    except:
        print(f"WARNING: Qubit #{i} failed to fit in the draft round. Skipped this qubit.")
        
    return "Qubit #{i} Failed."

    
    
```
